```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.express as px
```

```python
redwine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';', header=0)
redwine.head()
```

```python
# 두 데이터 셋을 합치기 위하여 type 지정
redwine['type'] = 0 
redwine.head()
```

```python
whitewine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv', sep=';', header=0)
whitewine.head()
```

```python
# 두 데이터 셋을 합치기 위하여 type 지정
whitewine['type'] = 1
whitewine.head()
```

```python
wine = pd.concat([redwine, whitewine], axis=0)  # 행으로 추가 
wine.columns = wine.columns.str.replace(' ', '_')   # columns의 공백을 _ 로 replace 

print(wine.info())
```

```python
#전체 데이터를 학습과 테스트 데이터셋으로 분리
from sklearn.model_selection import train_test_split

#X(입력변수)와 y(종속변수)로 이루어진 데이터를 학습과 테스트 데이터넷으로 각각 70%, 30%의 비율로 나눔 
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3, random_state=1, stratify=y) 
```

```python
# 표준화    # 정규화를 수행할 경우 회귀계수 간 상대적 중요도에 대한 올바른 해석을 위해 표준화가 필요함  
from sklearn.preprocessing import StandardScaler        # StandardScaler 표준화 진행
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y) 

# 데이터 표준화
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

# KNN 모델 학습
knn = KNeighborsClassifier(n_neighbors=5, p=2)  # KNN 모델 생성
knn.fit(X_train_std, y_train)                   # 모델 학습

# 테스트 데이터에 대한 예측
y_train_pred = knn.predict(X_train_std)         # 학습 데이터 예측값
y_test_pred = knn.predict(X_test_std)           # 테스트 데이터 예측값

# 정확도 계산 및 출력
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print("학습 데이터 정확도:", train_accuracy)
print("테스트 데이터 정확도:", test_accuracy)
```

```python
n_neighbors_values = [2,3,5,7]

for n_neighbors in n_neighbors_values:

    # KNN 모델 학습
    knn = KNeighborsClassifier(n_neighbors=n_neighbors, p=2)  # KNN 모델 생성
    knn.fit(X_train_std, y_train)                   # 모델 학습

    # 테스트 데이터에 대한 예측
    y_train_pred = knn.predict(X_train_std)         # 학습 데이터 예측값
    y_test_pred = knn.predict(X_test_std)           # 테스트 데이터 예측값

    # 정확도 계산 및 출력
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)

    print("{}의 정확도".format(n_neighbors))
    print("학습 데이터 정확도:", train_accuracy)
    print("테스트 데이터 정확도:", test_accuracy, '\n')
```

```python
n_neighbors_values = [2,3,5,7]

for n_neighbors in n_neighbors_values:

    # KNN 모델 학습
    knn = KNeighborsClassifier(n_neighbors=n_neighbors, p=1)  # KNN 모델 생성
    knn.fit(X_train_std, y_train)                   # 모델 학습

    # 테스트 데이터에 대한 예측
    y_train_pred = knn.predict(X_train_std)         # 학습 데이터 예측값
    y_test_pred = knn.predict(X_test_std)           # 테스트 데이터 예측값

    # 정확도 계산 및 출력
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)

    print("{}의 정확도".format(n_neighbors))
    print("학습 데이터 정확도:", train_accuracy)
    print("테스트 데이터 정확도:", test_accuracy, '\n')
```

```python
#분류나무(Classification Tree)
from sklearn.tree import DecisionTreeClassifier 

# 모델학습
dtc = DecisionTreeClassifier(criterion='gini', max_depth=7, random_state=1)     #깊이 max_depth 너무 깊이가 깊으면 과적합 스스로가 알아서 해야함 / 계수는 이제 본인이 알아서 
dtc.fit(X_train, y_train)

# 테스트 데이터 예측
y_train_pred = dtc.predict(X_train)  #Training accuracy
y_test_pred = dtc.predict(X_test)    #Test accuracy

# 정확도
print(accuracy_score(y_train, y_train_pred))
print(accuracy_score(y_test, y_test_pred))
```
